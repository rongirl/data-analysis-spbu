```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
def print_metrics(y_test, y_pred):
    print(f"MAE:  {round(mean_absolute_error(y_test, y_pred), 4)}")
    print(f"MSE:  {round(mean_squared_error(y_test, y_pred, squared=True), 4)}")
    print(f"RMSE: {round(mean_squared_error(y_test, y_pred, squared=False), 2)}")
    print(f"MPE:  {round(np.average(((y_test - y_pred) / y_test)) * 100, 4)}")
    print(f"MAPE: {round(np.average(((y_test - y_pred) / y_test).abs()) * 100, 4)}")
    print(f"AD:   {round(np.abs(y_pred - np.average(y_test)).sum(), 4)}")
    print(f"MAD:  {round(np.average(np.abs(y_pred - np.average(y_test))), 2)}")
    print(f"R^2:  {round(r2_score(y_test, y_pred), 2)}")
    print(f"v:    {round((((y_test - y_pred)**2).sum() / ( (y_test**2).sum() + (y_pred**2).sum() ))**0.5, 2)}")
```

```python
df = pd.read_csv('flat_prices.csv', index_col=0)
df.head()
```

```python
x = df[['SQUARE_METERS']]
y = df['PRICE']
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.25, random_state=1, shuffle=True)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
print_metrics(y_test, y_pred)
```

```python
plt.scatter(x_train['SQUARE_METERS'], y_train, color='purple', label='Train')
plt.scatter(x_test['SQUARE_METERS'], y_test, color='yellow', label='Test')
plt.plot(x_test['SQUARE_METERS'], y_pred, color='green', label='Regression Line')
plt.xlabel('Square Meters')
plt.ylabel('Price')
plt.legend()
plt.show()
```
