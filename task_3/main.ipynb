```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import warnings 
```

```python
warnings.filterwarnings("ignore")
```

```python
df = pd.read_csv("Electric_Production.csv")
#df['DATE'] = pd.DatetimeIndex(df['DATE']).year  
df['DATE'] = pd.to_datetime(df['DATE'])
df = df.rename(columns={'DATE': 'Date','IPG2211A2N': 'Electric Production'})
df = df.set_index(['Date'])
```

```python
df
```

```python
plt.figure(figsize = (10,6))
plt.plot(df['Electric Production'], color = 'cornflowerblue')
plt.xlabel('Date', size = 12)
plt.ylabel('Electric Production', size = 12)
plt.show()
```

```python
plt.figure(figsize = (10,6))
plt.hist(df['Electric Production'], color = 'cornflowerblue')
plt.xlabel('Electric Production', size = 12)
plt.ylabel('Frequency', size = 12)
plt.show()
```
